```python
import matplotlib. pyplot as plt
import pandas as pd
df=pd.read_csv('https://cocl.us/datascience_survey_data',index_col=0)
df.sort_values(by='Very interested',ascending=False,axis=0,inplace=True)
df=round(df/2233,2)
df.plot(kind='bar',figsize=(20,8),width=0.8,color=['#5cb85c','#5bc0de','#d9534f'],fontsize=14,)
plt.legend(['very interested','somewhat interested','not interested'],fontsize=14)
plt.title('Percentage of Respondants interested in data scince',fontsize=16)
plt.tick_params(labelleft='off',labelbottom='on')
```

```python
import pandas as pd
df=pd.read_csv('https://cocl.us/sanfran_crime_dataset',index_col=0)
df.drop(['Category','Descript','DayOfWeek','Date','Time','Resolution','Address','X','Y','Location'],axis=1,inplace=True)
df.columns=['Neighbourhood','Count']
dfs=df.groupby('Neighbourhood',axis=0,sort=True).count()
dfs.reset_index()
```

```python
import os
import pandas as pd
import folium
from folium.features import Choropleth
SF_map=folium.Map(location=[37.7749,122.4194],zoom_start=12,tiles='Mapbox  Bright')
SF_geo='san-francisco.geojson'
SF_map.choropleth(
Geo_path=SF_geo,
Data=df,
Columns=['Neighbourhood','Count'],
Key_on='feature.properties.name',
Fill_color='yl0rRd',
fill_opacity=0.7,
line_opacity=0.2,
    Legend_name='Crime rate in San Francisco'
)
SF_map
```
